Shootouts Analysis

```python
shootouts = pd.read_csv('data/shootouts.csv')
```

```python
shootouts.head()
```

```python
shootouts.isnull().sum()
```

```python
shootouts['first_shooter'].isnull().sum()
```

```python
# Most of the first_shooter values are missing, dropping this column is not a good idea
# Will fill missing values randomly between the home_team and away_team
import numpy as np

def fill_random(row):
    if pd.isnull(row['first_shooter']):
        return np.random.choice([row['home_team'], row['away_team']])
    else:
        return row['first_shooter']

shootouts['first_shooter'] = shootouts.apply(fill_random, axis=1)
# Axis 1 means we are applying the function to each row
```

```python
shootouts['first_shooter'].isnull().sum()
```

```python
shootouts['date'] = pd.to_datetime(shootouts['date'], format="%Y-%m-%d")
```

```python
shootouts.shape
```

```python
shootouts.isnull().sum()
```

```python
shootouts.dtypes
```

```python
shootouts.shape
```

```python
shootouts['first_shooter_won'] = (shootouts['first_shooter'] == shootouts['winner'])
win_counts = shootouts['first_shooter_won'].value_counts().reset_index()
win_counts.columns = ['lost/won', 'count']
win_counts
# what percentage of teams won by being the first shooters
```
